```python
import pandas as pd
import matplotlib.pyplot as plt
import os

files = [i for i in os.listdir() if '.csv' in i]

fig, ax = plt.subplots(2,2, figsize = (15,15))
for i,j in zip(files,ax.flatten()):
    data = pd.read_csv(i)
    votes = data['nVotes']
    labels = data[data.columns[0]]
    title = i.split('.')[0]
    j.pie(votes, labels = labels, autopct='%1.1f%%')
    j.title.set_text(title)
```

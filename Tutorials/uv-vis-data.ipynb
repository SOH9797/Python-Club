````python
def clean_data(path):
    df = pd.read_csv(path) # read file
    df = df.iloc[1:,:] # trim the first row out
    df.index = df.iloc[:,0] # set the index to wavelengths
    df = df.iloc[:,1::2].dropna() # drop rows with empty cells (tail info)
    df.columns = headers[::2][:-1] # every other header, except the last one
    df = df.astype(float) # some numbers are strings 
    df = df - df.iloc[0,:] # zero @ 800
    df = df.subtract(df['baseline'], axis=0) # subtract buffer baseline
    return df

df = clean_data('tutorial-data/titrations/20191128_ArachadonicaAcid_BM3WT_Titration_2.csv')
df
````

# extracting the compound concentrations
1. extract columns that contain ```"Arachadonic_acid"```
2. from those, extract numbers 
3. change number datatype from ```str``` to ```float```
4. dilution calculation

````python
import re

def substrate_vols(cols):
    substrate = [i for i in df.columns if 'Arachadonic_acid' in i]
    regex = [re.findall('\d+.\d+', i) for i in substrate] # add some cheat sheet cmds
    vols = [float(i[0]) for i in regex]
    return vols

vols = substrate_vols(df.columns)
vols
````

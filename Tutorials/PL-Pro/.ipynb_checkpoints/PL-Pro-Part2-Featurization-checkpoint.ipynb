```python
import pandas as pd

df = pd.read_csv('AID_1706.csv', index_col=0)

print(len(df))
df.head()
```

```python
import matplotlib.pyplot as plt

inactive = df.loc[df['PUBCHEM_ACTIVITY_OUTCOME']=='Inactive']
active = df.loc[df['PUBCHEM_ACTIVITY_OUTCOME']=='Active']

plt.figure(figsize=(7,5))
plt.scatter(inactive['Inhibition'], inactive['PUBCHEM_ACTIVITY_SCORE'], s=5, alpha = 0.5)
plt.scatter(active['Inhibition'], active['PUBCHEM_ACTIVITY_SCORE'],s=5,alpha = 0.5)
plt.legend(['Inactive','Active'])
plt.xlabel('Inhibition Metric')
plt.ylabel('Pubchem Activity Score')
plt.grid()
plt.show()
```

```python
from rdkit import Chem
from rdkit.Chem import PandasTools

df_sample = df.sample(1000)
PandasTools.AddMoleculeColumnToFrame(df_sample, 
                                     smilesCol = 'smiles',
                                     molCol='Structure')

df_sample[['PUBCHEM_ACTIVITY_OUTCOME', 'PUBCHEM_ACTIVITY_SCORE','Structure']][0:5]
```

```python
from rdkit.Chem import Crippen, AllChem

def PhysChemFeatures(smiles):
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    dictionary = {'MolWT':AllChem.CalcExactMolWt(mol),
                  'LogP':Chem.Crippen.MolLogP(mol),
                  'HBondDonors': AllChem.CalcNumLipinskiHBD(mol),
                  'HBondAcceptors': AllChem.CalcNumLipinskiHBA(mol)}
    
    return pd.Series(dictionary)

df.sample(1000)['smiles'].apply(PhysChemFeatures).head()
```

```python
%%timeit
df.sample(1000)['smiles'].apply(PhysChemFeatures)
```

```python
%%timeit
from concurrent.futures import ThreadPoolExecutor

with ThreadPoolExecutor(max_workers=1) as executor:
    future = executor.map(PhysChemFeatures, df.sample(1000)['smiles'])
                          
result = pd.DataFrame(list(future))
```

```python
%%timeit
from multiprocessing import Pool, cpu_count

pool = Pool(processes = (cpu_count() - 1))

results = pool.map(PhysChemFeatures, df.sample(1000)['smiles'])
pool.close()
pool.join()
pd.DataFrame(results)
```

```python
from multiprocessing import Pool, cpu_count
import time

pool = Pool(processes = (cpu_count() - 1))

start_timer = time.time()
results = pool.map(PhysChemFeatures, df['smiles'])
pool.close()
pool.join()
stop_timer = time.time()

results = pd.DataFrame(results)
print(f'{round(stop_timer - start_timer)} s to apply function to {len(df)} cells')
results.head()
```

```python
lipinski_friendly = results.loc[results['MolWT']<500]\
.loc[results['LogP']<5]\
.loc[results['HBondDonors']<5]\
.loc[results['HBondAcceptors'] <10]

not_lipinski = results.drop(lipinski_friendly.index)

print(f"Percentage of compounds that conform to Lipinski's rule of 5: \
{100* round(len(lipinski_friendly)/len(results), 3)} %")
```

```python
import matplotlib.pyplot as plt

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharey=True, figsize = (10,8))
fig.suptitle('Distibution of Lipinski Descriptors in Screening Compounds')

for i,j in zip([ax1, ax2, ax3, ax4], results.columns):
    i.hist(results[j], bins = 50)
    i.set_xlabel(j)
    i.set_ylabel('Count')
plt.show()
```

```python
PandasTools.AddMurckoToFrame(df_sample, molCol = 'Structure', MurckoCol='Scaffold smiles')
PandasTools.AddMoleculeColumnToFrame(df_sample, smilesCol='Scaffold smiles',molCol='Scaffold')
df_sample[['PUBCHEM_ACTIVITY_OUTCOME', 'PUBCHEM_ACTIVITY_SCORE','Structure','Scaffold']][0:5]
```

```python
from rdkit.Chem.Scaffolds.MurckoScaffold import MurckoScaffoldSmilesFromSmiles


start_timer = time.time()

pool = Pool(processes = (cpu_count() - 1))
scaffolds = pool.map(MurckoScaffoldSmilesFromSmiles, df['smiles'])
pool.close()
pool.join()

stop_timer = time.time()

scaffolds = pd.Series(scaffolds)
print(f'{round(stop_timer - start_timer)} s to apply function to {len(df)} cells')
scaffolds.head()
```

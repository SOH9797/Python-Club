````python
def clean_data(filename):
    data = pd.read_csv(filename, skiprows=5)
    df = data.loc[1:,:]
    df.columns = data.loc[0,:] 
    df.index = [i+str(int(j)) for i,j in zip(df.iloc[:,0], df.iloc[:,1])]
    df = df.loc[:, (cols.dropna() == False).to_list()]
    df.columns = [int(i) for i in df.columns]


    return df

df = clean_data('tutorial-data/20200208_pilot_1YQO.CSV')
df
````

````python
from string import ascii_uppercase
top = [[i + str(j) for i in ascii_uppercase[0:8]] for j in range(1,25)] 
bottom = [[i + str(j) for i in ascii_uppercase[8:16]] for j in range(1,25)] 

print(top)
print()
print(bottom)
````

````python
blocks = [[top[i], top[i+1]] for i in range(0,24,2)]
for i in [[bottom[i], bottom[i+1]] for i in range(0,24,2)]:
    blocks.append(i)
blocks
````

````python
first_block = blocks[0]
first_block
````

````python
test = df.loc[first_block[0],:]
ctrl = df.loc[first_block[1],:]
````

````python
def trace(x):
    plt.figure(figsize=(12,4))
    for i in x.index:
        plt.plot(x.columns, x.loc[i,:])
    plt.xlabel('Wavelength - nm')
    plt.ylabel('Absorbance')
    plt.ylim(0,1)
    plt.xlim(250,800)
    plt.show()
trace(test)
````

````python
trace(ctrl)
````

````python
def zero(x):
    return x.subtract(x[799], axis =0)
zero(test)
````

````python
trace(zero(test))
````
